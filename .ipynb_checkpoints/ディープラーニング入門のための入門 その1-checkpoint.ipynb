```python
#!/usr/bin/env python
# coding:utf-8
"""
[1,2,3]というベクトルを与え,中の数字をそれぞれ2倍にして返すモデルを作りたい
"""
import numpy as np
import chainer.functions as F
from chainer import Variable, FunctionSet, optimizers

#モデル定義(線形関数の指示だけで,中の数字をそれぞれ2倍にしたいとは指示していない.)
model = F.Linear(3,3) # 何をやっているのか? class Linear(function.Function): の(3.3)?

#モデルに与えるベクトル[1,2,3]
data = np.array([[1,2,3]],dtype=np.float32)

x = Variable(data) # Variable グローバル変数?

#ここでモデルにベクトルを投げ,加工
y = model(x)

#モデルが加工したベクトルを表示
print(y.data)
```

```python
#!/usr/bin/env python
# coding:utf-8
"""
[1,2,3]というベクトルを与え、中の数字をそれぞれ2倍になるよう,50回繰り返す
上記のcodeと同じところに#をつける.
"""
import numpy as np #
import chainer.functions as F #
from chainer import Variable, FunctionSet, optimizers #

#モデル定義
model = F.Linear(3,3) #
optimizer = optimizers.SGD() 
optimizer.setup(model) 


#学習させる回数
times = 50


#与えるベクトル  これが[2,4,6]になって返ってきて欲しい
x = Variable(np.array([[1,2,3]], dtype=np.float32)) #


#正解ベクトルの[2,4,6]
t = Variable(np.array([[2,4,6]], dtype=np.float32))

#ここから50回ループ
for i in range(0,times):
    optimizer.zero_grads()

    #ここでモデルにベクトルを投げ,加工
    y = model(x)

    #モデルが出した答えを表示
    print(y.data)

    #モデルが出した答えと,本当の答え([2,4,6])がどのくらい違っているか計算する
    loss = F.mean_squared_error(y, t)

    #その値をモデルに見せて「全然違う.もっと近づけろ」と学習させる
    loss.backward()
    optimizer.update()

    #最初に戻って繰り返す
```
